```python
diccionario = {}

while True:
    print("\n---MENU---") 
    print("1. Crear datos nuevos")
    print("2. Modificar datos")
    print("3. Elimiar Datos")
    print("4. Salir")
    
    opcion = input("Elige una opcion (1-4)")

    if opcion == "1":
        clave = input ("Introduce el dato")
        valor = input (f"introduce el valor para {clave}:")
        diccionario[clave] = valor
        print(f"dato cargado: {clave} = {valor} ")

    elif opcion == "2":
        clave = input ("que dato deseas modificar?:")
        if clave in diccionario:
            nuevo_valor = input (f"introduce la modificacion para {clave}")
            diccionario[clave] = nuevo_valor
            print(f"Dato modificado: {clave} = {nuevo_valor}")
        else:
            print(f"No existe el dato '{clave}' en la tarjeta")    

    elif opcion == "3":
        clave = input ("que dato deseas eliminar?:")
        if clave in diccionario:
            del diccionario[clave]
            print(f"dato eliminado: {clave}")
        else:
            print(f"No existe el dato '{clave}' en la tarjeta")
   
    elif opcion == "4":
        print ("saliendo del sistema")
        break
    else:
        print("opcion no valida. Intenta de nuevo")  
 
  
```

```python
agenda = {}

while True:
    print("\n--- AGENDA DE CONTACTOS ---")
    print("1. Crear contacto")
    print("2. Mostrar los contactos")
    print("3. Modificar un contacto")
    print("4. Salir")

    opcion = input("Elige una opcion (1-4):")


    if opcion == 1:
        nombre = input("nombre del contacto:")
        if nombre in agenda:
            print("El contacto ya existe")
        else:
            telefono = input("telefono:")
            correo = input("correo:")
            agenda[nombre] = {"telefono": telefono, "correo": correo}
            print("contacto agregado exitosamente")
   

    elif opcion == "2":
        if not agenda:
            print("la agenda esta vacia")
        else:
            print ("\n--- LISTA DE CONTACTOS ---")    
            for nombre, datos in agenda.items():
                print(f"Nombre: {nombre}")
                print(f"  Teléfono: {datos['Teléfono']}")
                print(f"  Correo: {datos['Correo']}")
   

    elif opcion == "3":
        nombre = input("Introduce el nombre del contacto a modificar: ")
        if nombre in agenda:
            print("Datos actuales:")
            print(f"  Teléfono: {agenda[nombre]['Teléfono']}")
            print(f"  Correo: {agenda[nombre]['Correo']}")
            
            nuevo_telefono = input("Nuevo teléfono (dejar vacío si no se modifica): ")
            nuevo_correo = input("Nuevo correo (dejar vacío si no se modifica): ")
            
            if nuevo_telefono:
                agenda[nombre]["Teléfono"] = nuevo_telefono
            if nuevo_correo:
                agenda[nombre]["Correo"] = nuevo_correo

            print("Contacto modificado exitosamente.")
        else:
            print("El contacto no existe en la agenda.")      


    elif opcion == "4":
        print("Saliendo del programa...")
        break

    else:
        print("Opción inválida. Intenta de nuevo.")      




```
